```python
# 先降维再分组
import warnings

warnings.filterwarnings(action='ignore')
import xgboost as xgb
import pandas as pd
import numpy as np
from sklearn.model_selection._validation import cv_results_fth
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, KFold
from sklearn.metrics import precision_score, roc_curve, recall_score, f1_score, roc_auc_score, accuracy_score
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.svm import SVR, SVC
from sklearn import svm
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
# from sklearn.feature_selection import SequentialFeatureSelector as SFS
from sklearn.feature_selection import VarianceThreshold
from sklearn.feature_selection import SelectKBest
from scipy.stats import pearsonr
from sklearn.linear_model import LassoCV, LogisticRegression, Lasso, LassoLarsCV, MultiTaskLasso, MultiTaskLassoCV
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
from sklearn.ensemble import RandomForestClassifier
import shap
shap.initjs()


def dataprocessing(data):
    dummies_Sex = pd.get_dummies(data['Sex'], prefix='Sex')
    dummies_Sex = pd.DataFrame(dummies_Sex)

    ECOG = {'ECOG 0': 0, 'ECOG 1': 1, 'ECOG 2': 2, 'ECOG 3': 3, 'ECOG 4': 4}
    data["ECOG_Performance"] = data["ECOG_Performance"].map(ECOG)

    smoker = {'Non-smoker': 0, 'Ex-smoker': 1, 'Current': 2}
    data["Smoking"] = data["Smoking"].map(smoker)

    drinker = {'non-drinker': 0, 'light': 1, 'Moderate': 2, 'ex-drinker': 3, 'heavy': 4, 'unknown': 5}
    data["Drinking history"] = data["Drinking history"].map(drinker)

    sub = {'post wall': 1, 'Tonsillar Fossa': 2, 'Base of Tongue': 3, 'Tonsil ': 4,
           'Tonsil Pillar ': 5, 'Soft Palate ': 6, 'Vallecula': 7, 'lat wall': 0}
    data["Subsite"] = data["Subsite"].map(sub)

    T = {'T1': 1, 'T2': 2, 'T3': 3, 'T3 (2) ': 4, 'T4a': 5, 'T4b': 0}
    data["T"] = data["T"].map(T)

    N = {'N0': 0, 'N1': 1, 'N2a': 2, 'N2b': 3, 'N2c': 4, 'N3': 5}
    data["N"] = data["N"].map(N)

    M = {'M0': 0}
    data["M"] = data["M"].map(M)

    stage = {'I': 0, 'II': 1, 'III': 2, 'IVA': 3, 'IVB': 4}
    data["Stage"] = data["Stage"].map(stage)

    Hpv = {'  Negative': 0, '  positive': 1}
    data["HPVp16status"] = data["HPVp16status"].map(Hpv)

    Che = {'none': 0, 'Yes': 1}
    data["Chemotherapy"] = data["Chemotherapy"].map(Che)

    RT = {'IMRT': 0, 'IMRT-ipsilat': 1}
    data["RT_Tech"] = data["RT_Tech"].map(RT)

    Status = {'Dead': 0, 'Alive': 1}
    data["Status"] = data["Status"].map(Status)

    data.drop(['PatientID', 'Sex', 'Ds_Site', 'Path', 'Primary_Treatment', 'Cause_of_Death',
               'Local_Failure', 'local_failure(days)', 'Regional_Failure', 'regional_failure(days)',
               'Distant_Failure', 'distant_failure (days)', 'Second_Primary', '2nd_cancer_(days)',
               'PMID', 'diagnostics_Versions_PyRadiomics', 'diagnostics_Versions_Numpy',
               'diagnostics_Versions_SimpleITK', 'diagnostics_Versions_PyWavelet', 'diagnostics_Versions_Python',
               'diagnostics_Configuration_Settings', 'diagnostics_Configuration_EnabledImageTypes',
               'diagnostics_Image-original_Hash', 'diagnostics_Image-original_Dimensionality',
               'diagnostics_Image-original_Spacing', 'diagnostics_Image-original_Size',
               'diagnostics_Image-original_Mean',
               'diagnostics_Image-original_Minimum', 'diagnostics_Image-original_Maximum',
               'diagnostics_Mask-original_Hash', 'diagnostics_Mask-original_Spacing',
               'diagnostics_Mask-original_Size',
               'diagnostics_Mask-original_BoundingBox', 'diagnostics_Mask-original_VoxelNum',
               'diagnostics_Mask-original_VolumeNum', 'diagnostics_Mask-original_CenterOfMassIndex',
               'diagnostics_Mask-original_CenterOfMass'

               ], axis=1, inplace=True)

    data_df = pd.concat([data, dummies_Sex], axis=1)
    return data_df

# 交叉评估
def cross_validation_test(x_train, y_train, x_test, y_test):
    model_xgb = xgb.XGBClassifier()
    cv_results = cv_results_fth(model_xgb, x_train, y_train, cv=5, scoring='roc_auc')
    # print(cv_results)
    models = cv_results['estimator']
    print("cv验证结果：")
    print(cv_results['test_score'])
    # print(np.max(cv_results['test_score']))
    print((cv_results['test_score']).mean())
    print("独立测试：")
    auc_result = []
    for model in models:
        y_pre = model.predict_proba(x_test)
        y_pre_1 = y_pre[:, 1]
        org_auc = roc_auc_score(y_test, y_pre_1)
        auc_result.append(org_auc)
        print(org_auc)
    print("均值：", np.mean(auc_result))
    return cv_results

# 非交叉评估
def no_cross_validation_test(x_train, y_train, x_test, y_test):
    model_xg = xgb.XGBClassifier().fit(x_train, y_train)
    print("非交叉评估test:")
    y_pre = model_xg.predict_proba(x_test)
    y_pre_1 = y_pre[:, 1]
    org_auc = roc_auc_score(y_test, y_pre_1)
    print(org_auc)


# 绘制auc
def ks(y_predicted1, y_true1, y_predicted2, y_true2, y_predicted3, y_true3):
  Font={'size':18, 'family':'Times New Roman'}
  
  label1=y_true1
  label2=y_true2
  label3=y_true3
  fpr1,tpr1,thres1 = metrics.roc_curve(label1, y_predicted1)
  fpr2,tpr2,thres2 = metrics.roc_curve(label2, y_predicted2)
  fpr3,tpr3,thres3 = metrics.roc_curve(label3, y_predicted3)
  roc_auc1 = metrics.auc(fpr1, tpr1)
  roc_auc2 = metrics.auc(fpr2, tpr2)
  roc_auc3 = metrics.auc(fpr3, tpr3)
  
  plt.figure(figsize=(6,6))
  plt.plot(fpr1, tpr1, 'b', label = 'Stacking = %0.3f' % roc_auc1, color='Red')
  plt.plot(fpr2, tpr2, 'b', label = 'XGBoost = %0.3f' % roc_auc2, color='k')
  plt.plot(fpr3, tpr3, 'b', label = 'Random Forest = %0.3f' % roc_auc3, color='RoyalBlue')
  plt.legend(loc = 'lower right', prop=Font)
  plt.plot([0, 1], [0, 1],'r--')
  plt.xlim([0, 1])
  plt.ylim([0, 1])
  plt.ylabel('True Positive Rate', Font)
  plt.xlabel('False Positive Rate', Font)
  plt.tick_params(labelsize=15)
  plt.show()
  return abs(fpr1 - tpr1).max(),abs(fpr2 - tpr2).max(),abs(fpr3 - tpr3).max()

# 处理数据

noLightSpeedPlus = "./opc_data/noLightSpeedPlus.csv"
train_data = pd.read_csv(noLightSpeedPlus)
print(train_data.shape)

LightSpeedPlus = "./opc_data/LightSpeedPlus.csv"
independent_test_data = pd.read_csv(LightSpeedPlus)
print(independent_test_data.shape)

# total_data = "./opc_data/train_data_519.csv"
total_data = "./opc_data/LightSpeedPlus.csv"

total_data = pd.read_csv(total_data)

dataprocessing(train_data)
dataprocessing(independent_test_data)
dataprocessing(total_data)
print(train_data.shape,independent_test_data.shape)

train_data = train_data.fillna(0)
independent_test_data = independent_test_data.fillna(0)
total_data = total_data.fillna(0)


y_train = train_data["Status"]
x_train = train_data.drop(["FU", "Status"], axis=1)
# print(y_train)
# print(x_train)
y_test = independent_test_data["Status"]
x_test = independent_test_data.drop(["Status", "FU"], axis=1)

y_total = total_data["Status"]
x_total = total_data.drop(["FU", "Status"], axis=1)

x_train = x_train.astype(np.float64)
x_test = x_test.astype(np.float64)
x_total = x_total.astype(np.float64)


x_nost_train = x_train
x_nost_test = x_test

# 标准化
x_columns = x_train.columns
stScale = StandardScaler().fit(x_train)
x_train = stScale.transform(x_train)
x_test = stScale.transform(x_test)

x_train = pd.DataFrame(x_train)
x_train.columns = x_columns
x_test = pd.DataFrame(x_test)
x_test.columns = x_columns

stScale = StandardScaler().fit(x_total)
x_total = stScale.transform(x_total)
x_total = pd.DataFrame(x_total)
x_total.columns = x_columns

# 降维前
# 使用五折交叉验证
no_cross_validation_test(x_train, y_train, x_test, y_test)
cv_results_orginal = cross_validation_test(x_train, y_train, x_test, y_test)

# lasso降维
# alphas = np.logspace(-4, -2, 50)
# # print(alphas)
# model_lassoCV = LassoCV(alphas=alphas, cv=10, max_iter=1000, n_jobs=-1).fit(x_total, y_total)
# print(model_lassoCV.coef_)
# print("model_lassoCV.alpha_", model_lassoCV.alpha_)
# coef_Status = pd.Series(model_lassoCV.coef_, index=x_total.columns)
# print(coef_Status)
# print( "Lasso picked " + str(sum(coef_Status != 0)) + " variables and eliminated the other " + str(sum(coef_Status == 0)))
#
# # 用features存储lasso降维后的特征
# features = coef_Status[coef_Status != 0].index
# print(features)

features = ['Age at diagnosis', 'ECOG_Performance', 'Smoking pack year', 'Drinking history', 'T',
       'N', 'Stage', 'HPVp16status', 'Chemotherapy', 'Dose',
       'original_shape_Maximum2DDiameterRow',
       'original_shape_Maximum2DDiameterSlice',
       'original_shape_Maximum3DDiameter', 'original_glrlm_RunVariance',
       'wavelet-LLH_glcm_Correlation', 'wavelet-LLH_glcm_Imc1',
       'wavelet-LLH_glcm_Idn',
       'wavelet-LLH_glszm_LargeAreaHighGrayLevelEmphasis',
       'wavelet-LHL_firstorder_Skewness', 'wavelet-LHL_glcm_ClusterShade',
       'wavelet-LHL_glcm_Imc1', 'wavelet-LHL_glcm_Idmn',
       'wavelet-LHL_gldm_DependenceEntropy', 'wavelet-LHH_firstorder_Skewness',
       'wavelet-LHH_glcm_ClusterShade', 'wavelet-LHH_glcm_InverseVariance',
       'wavelet-LHH_glszm_GrayLevelNonUniformityNormalized',
       'wavelet-LHH_ngtdm_Busyness', 'wavelet-LHH_gldm_DependenceVariance',
       'wavelet-HLL_glszm_SmallAreaEmphasis', 'wavelet-HLL_ngtdm_Coarseness',
       'wavelet-HLL_gldm_LargeDependenceLowGrayLevelEmphasis',
       'wavelet-HLH_firstorder_Median', 'wavelet-HLH_firstorder_Skewness',
       'wavelet-HHL_firstorder_Mean',
       'wavelet-HHL_glszm_GrayLevelNonUniformityNormalized',
       'wavelet-HHH_glrlm_LongRunLowGrayLevelEmphasis',
       'wavelet-HHH_glszm_LargeAreaLowGrayLevelEmphasis',
       'wavelet-LLL_firstorder_Kurtosis', 'wavelet-LLL_firstorder_Median',
       'wavelet-LLL_ngtdm_Contrast', 'exponential_firstorder_Kurtosis',
       'exponential_glcm_InverseVariance',
       'exponential_glszm_SmallAreaHighGrayLevelEmphasis',
       'exponential_ngtdm_Coarseness', 'gradient_firstorder_Minimum',
       'gradient_glcm_Idmn', 'gradient_glszm_SmallAreaEmphasis',
       'gradient_gldm_DependenceNonUniformityNormalized',
       'lbp-3D-m1_glszm_GrayLevelNonUniformity',
       'lbp-3D-m1_glszm_SizeZoneNonUniformityNormalized',
       'lbp-3D-m1_glszm_ZoneEntropy', 'lbp-3D-m1_glszm_ZoneVariance',
       'lbp-3D-m2_firstorder_Skewness', 'lbp-3D-m2_glszm_ZoneVariance',
       'lbp-3D-k_firstorder_Minimum', 'lbp-3D-k_firstorder_RootMeanSquared',
       'lbp-3D-k_glszm_HighGrayLevelZoneEmphasis',
       'lbp-3D-k_glszm_SmallAreaLowGrayLevelEmphasis',
       'lbp-3D-k_gldm_LargeDependenceHighGrayLevelEmphasis',
       'logarithm_firstorder_InterquartileRange',
       'logarithm_firstorder_Kurtosis', 'logarithm_glcm_Imc2',
       'logarithm_glszm_GrayLevelNonUniformity',
       'logarithm_gldm_LargeDependenceLowGrayLevelEmphasis',
       'square_firstorder_Kurtosis', 'square_glrlm_ShortRunEmphasis',
       'square_glszm_SmallAreaLowGrayLevelEmphasis',
       'square_gldm_LargeDependenceHighGrayLevelEmphasis',
       'squareroot_glrlm_GrayLevelNonUniformity', 'squareroot_ngtdm_Busyness']

x_train = x_train[features]
x_test = x_test[features]
x_total = x_total[features]
# 评估lasso选出来的特征集
no_cross_validation_test(x_train, y_train, x_test, y_test)
cv_results_lasso = cross_validation_test(x_train, y_train, x_test, y_test)

# 序列浮动算法
# model_xgb = xgb.XGBClassifier()
# sfs = SFS(model_xgb, k_features=(1, 71), forward=False, floating=True, scoring='roc_auc', cv=5, n_jobs=60)
# sfs = sfs.fit(x_total, y_total)
# # 显示所有列
# pd.set_option('display.max_columns', None)
# # 显示所有行
# pd.set_option('display.max_rows', None)
# print(pd.DataFrame.from_dict(sfs.get_metric_dict()).T)
# print('best combination (AUC: %.3f): %s\n' % (sfs.k_score_, sfs.k_feature_names_))
# plot_sfs(sfs.get_metric_dict(), kind='std_err')
# result_features = sfs.k_feature_names_
# result_score = sfs.k_score_
# # print("----------------------- subsets--------------------------------")
# # print(sfs.subsets_)

features_sfbs = ['Age at diagnosis', 'ECOG_Performance', 'Smoking pack year', 'Drinking history', 'HPVp16status', 'Chemotherapy', 'wavelet-LHH_firstorder_Skewness', 'wavelet-LHH_glszm_GrayLevelNonUniformityNormalized', 'wavelet-HHH_glrlm_LongRunLowGrayLevelEmphasis', 'lbp-3D-k_gldm_LargeDependenceHighGrayLevelEmphasis', 'logarithm_firstorder_Kurtosis', 'squareroot_glrlm_GrayLevelNonUniformity']
x_train = x_train[features_sfbs]
x_test = x_test[features_sfbs]
x_total = x_total[features_sfbs]

x_nost_train = x_nost_train[features_sfbs]
x_nost_test = x_nost_test[features_sfbs]
no_cross_validation_test(x_train, y_train, x_test, y_test)
cv_results_sfbs = cross_validation_test(x_train, y_train, x_test, y_test)

# shap解释
model_final=xgb.XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
              colsample_bynode=1, colsample_bytree=1, gamma=0, gpu_id=-1,
              importance_type='gain', interaction_constraints='',
              learning_rate=0.300000012, max_delta_step=0, max_depth=6,
              min_child_weight=1,monotone_constraints='()',
              n_estimators=150, n_jobs=8, num_parallel_tree=1, random_state=0,
              reg_alpha=0, reg_lambda=1, scale_pos_weight=1, subsample=1,
              tree_method='exact', validate_parameters=1, verbosity=None)
model_final.fit(x_nost_train,y_train)

result = model_final.predict_proba(x_nost_test)

explainer = shap.TreeExplainer(model_final, model_output="probability", feature_dependence="independence", data=x_nost_train)
print("mean of y_train",y_train.mean())
print("mean of y_test",y_test.mean())
print("mean of x_train_prediction",model_final.predict_proba(x_nost_train)[1].mean())
print("mean of x_test_prediction",model_final.predict_proba(x_nost_test)[1].mean())
print("treeshap-------------------------:",explainer.expected_value)
shap_values = explainer.shap_values(x_nost_test)
shap.summary_plot(shap_values, x_nost_test)
shap.summary_plot(shap_values, x_nost_test, plot_type="bar")

# print()
# print("treeshap:",explainer.expected_value)
# print("true",model_final.predict_proba(x_nost_test).mean())
# print("true",model_final.predict(x_nost_test).mean(0))
# result =  model_final.predict_proba(x_train)
# print(result.mean(0))
# print(np.mean(np.delete(result,0,axis = 1)))
# print(np.mean(np.delete(result,1,axis = 1)))


def shap_dependence_plot(features_name, shap_values, train_data):
    shap.dependence_plot(ind=features_name, shap_values=shap_values, features=train_data, interaction_index=None,
                         x_jitter=0, alpha=1)


for feature in features_sfbs:
    shap_dependence_plot(feature, shap_values, x_nost_test)
# print(x_nost_test.iloc[0:1])

# print(x_nost_test.iloc[1])

print(y_test, model_final.predict(x_nost_test))

display(shap.force_plot(explainer.expected_value, shap_values[2],x_nost_test.iloc[2], matplotlib=False))
shap.force_plot(explainer.expected_value, shap_values[2],x_nost_test.iloc[2],show=False,matplotlib=True).savefig('d://test.png',format = "png",dpi = 600,bbox_inches = 'tight')
# pre_0 = pd.DataFrame(shap_values[0],x_nost_test.columns).sort_values(by=0, ascending=False)
pre_2 = pd.DataFrame(shap_values[2],x_nost_test.columns)
display(pre_2)

feature_data = (x_nost_test.iloc[2]).to_frame()
# print(type(feature_data))
# print(x_nost_test.iloc[0])
case_2 = pd.concat([feature_data,pre_2], axis=1)
# display(case_0)
case_2.columns = ['feature values','shap values']
case_2 = case_2.sort_values(by="shap values", ascending=False)
display(case_2)



# shap.force_plot(explainer.expected_value, shap_values[3],x_nost_test.iloc[3],show=False,matplotlib=True).savefig('d://test.png')

# pre_0 = pd.DataFrame(shap_values[0],x_nost_test.columns).sort_values(by=0, ascending=False)
pre_3 = pd.DataFrame(shap_values[3],x_nost_test.columns)
display(pre_3)

feature_data = (x_nost_test.iloc[3]).to_frame()
# print(type(feature_data))
# print(x_nost_test.iloc[0])
case_3 = pd.concat([feature_data,pre_3], axis=1)
# display(case_0)
case_3.columns = ['feature values','shap values']
case_3 = case_3.sort_values(by="shap values", ascending=False)
display(case_3)
# display(shap.force_plot(explainer.expected_value, shap_values[16],x_nost_test.iloc[16],figsize=(50, 3), matplotlib=False))

# pre_16 = pd.DataFrame(shap_values[16],x_nost_test.columns)
# display(pre_16)

# feature_data = (x_nost_test.iloc[16]).to_frame()
# # print(type(feature_data))
# # print(x_nost_test.iloc[0])
# case_16 = pd.concat([feature_data,pre_16], axis=1)
# # display(case_0)
# case_16.columns = ['feature values','shap values']
# case_16 = case_16.sort_values(by="shap values", ascending=False)
# display(case_16)

# display(shap.force_plot(explainer.expected_value, shap_values[3],x_nost_test.iloc[3],figsize=(50, 3), matplotlib=False))
# display(pd.DataFrame(shap_values[3],x_nost_test.columns).sort_values(by=0, ascending=False))
```
